```python
from glob import glob
fns=glob('euler_nr_?.txt')
fns.sort()
```

```python
fig=plt.figure(figsize=(4,3))
ax=fig.add_subplot(111)
dt=[0.01,0.1,2]

for i in range(len(fns)):
    fn=fns[i]
    print('fn:',fn)
    dat=np.loadtxt(fn).T
    
    plot(dat[0],dat[1],'-',label=r'$\Delta t =$%4.2f'%dt[i])
    
ax.set_xlabel('time')
ax.set_ylabel('Length')
ax.legend()

fig.tight_layout()
fig.savefig('euler_nr_var_const_time_length.pdf',bbox_to_inches='tight')
```

```python
fig=plt.figure(figsize=(4,3))
ax=fig.add_subplot(111)
dt=[0.01,0.1,2]
for i in range(len(fns)):
    fn=fns[i]
    print('fn:',fn)
    dat=np.loadtxt(fn).T
    
    ax.plot(dat[0],dat[2],'-o',label=r'$\Delta t =$%4.2f'%dt[i])
    
ax.set_xlabel('time')
ax.set_ylabel('Force')
ax.legend()

fig.tight_layout()
fig.savefig('euler_nr_var_const.pdf',bbox_to_inches='tight')
```
