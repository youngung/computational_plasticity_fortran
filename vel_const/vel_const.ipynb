```python
def eps(t,l0,vel):
    return np.log(l0+vel*t)-np.log(l0)
def length(t,l0,vel):
    return l0+vel*t
```

```python
def elasticity(t,E,l0,c):
    return E*np.log((l0+c*t)/l0)
```

```python
def newtonian(t,eta,l0,c):
    return eta*c/(l0+c*t)
```

## Elasticity

```python
l0=20.
t=np.linspace(0.,30.)
vel=0.001
E=200e3
A=12.5 * 1. #![mm]


fig=plt.figure(figsize=(13,3))
ax1=fig.add_subplot(141)
ax3=fig.add_subplot(142)
ax4=fig.add_subplot(143)
ax5=fig.add_subplot(144)
ax2=ax1.twinx()



l=length(t,l0,vel)
e=eps(t,l0,vel)
s=elasticity(t,E,l0,vel)
f=A*s
ax1.plot(t,l,'-r',label='length')
ax2.plot(t,e,'--b',label='epsilon')
ax3.plot(t,s)
ax4.plot(e,s,'green')
ax5.plot(t,f/1e3,'orange')



## Deco

ax1.set_ylabel(r'Length [mm]')
ax1.set_xlabel('time [sec]')
ax1.legend(loc='upper left')
ax2.set_ylabel(r'$\varepsilon$')
ax2.legend(loc='lower right')
ax3.set_xlabel('time [sec]')
ax3.set_ylabel('Stress [MPa]')
ax4.set_xlabel('Strain')
ax4.set_ylabel('Stress [MPa]')
ax5.set_ylabel('Force [kN]')
ax5.set_xlabel('time [sec]')
fig.tight_layout()
fig.savefig('vel_cont_elast.pdf',bbox_inches='tight')
```

## Newtonian fluid

```python


fig=plt.figure(figsize=(13,3))
ax1=fig.add_subplot(141)
ax3=fig.add_subplot(142)
ax4=fig.add_subplot(143)
ax5=fig.add_subplot(144)
ax2=ax1.twinx()




eta = 1.3e-3 # waterat 10 Celcius degree 
l0=5.
A=12.5 * 1. #![mm]

t=np.linspace(0.,30.)
vel=-0.1
l=length(t,l0,vel)
e=eps(t,l0,vel)
f=A*s
s=newtonian(t,eta,l0,vel)
e=eps(t,l0,vel)

ax1.plot(t,l,'-r',label='length')
ax2.plot(t,e,'--b',label='epsilon')
ax3.plot(t,s)
ax4.plot(e,s,'green')
ax5.plot(t,f,'orange')




ax1.set_ylabel(r'Length [mm]')
ax2.set_ylabel(r'$\varepsilon$')
ax1.set_xlabel('time [sec]')
ax1.legend(loc='upper left')
ax2.legend(loc='lower right')
ax3.set_xlabel('time [sec]')
ax3.set_ylabel('Stress [Pa]')
ax4.set_xlim(0,-1)
ax4.set_ylim(0,-0.00007)
ax4.set_xlabel('Strain')
ax4.set_ylabel('Stress [MPa]')
ax5.set_ylabel('Force [N]')
ax5.set_xlabel('time [sec]')
fig.tight_layout()
fig.savefig('vel_cont_newtonian.pdf',bbox_inches='tight')
```
