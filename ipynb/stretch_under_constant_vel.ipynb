```python


def l(l0,t,sr):
    return l0*np.exp(sr*t)
```

```python
l0=100
t=np.linspace(0,100)

fig=plt.figure(figsize=(3,3))
ax=fig.add_subplot(111)
srs=[1e-3,2e-3,3e-3]
for sr in srs:
    ax.plot(t,l(l0,t,sr))
ax.set_xlabel('second [s]')
ax.set_ylabel('length [mm]')
ax.set_ylim(0,)
fig.savefig('l.png')
```
